```python
import pandas as pd
```

```python
movies = pd.read_csv("../data/movies.csv", index_col = "Title")
movies
```

```python
# extract a single column of data from a DataFrame as a Series.
movies["Studio"]
```

```python
# group by and count
movies["Studio"].value_counts()
```

```python
# Filtering a column by one criteria
movies[movies["Studio"] == "Universal"]
```

```python
# Filtering a column by more criterias(and)
released_by_universal = movies["Studio"] == "Universal" 
released_in_2015 = movies["Year"] == 2015 
movies[released_by_universal & released_in_2015]
```

```python
# Filtering a column by more criterias(or)
released_by_universal = movies["Studio"] == "Universal" 
released_in_2015 = movies["Year"] == 2015 
movies[released_by_universal | released_in_2015]
```

```python
before_1975 = movies["Year"] < 1975
movies[before_1975]
```
