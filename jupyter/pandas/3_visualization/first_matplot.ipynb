```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r"../data/Countries Population from 1995 to 2020.csv")
dataset
```

```python
# 筛选国家为中国的数据
china_pop_info = dataset[dataset.Country=="China"]
china_pop_info
```

```python
#### 画图 ####
# 传入x轴数据，y轴数据
plt.plot(china_pop_info['Year'], china_pop_info['Population'])
# 显示画图结果
plt.show() 
```

```python
# 从Country列统计一下一共有多少个不同的国家名字
dataset.Country.unique() 
```

```python
# draw plot of united states
us_pop_info = dataset[dataset.Country=="United States"]
plt.plot(us_pop_info['Year'], us_pop_info['Population'])
plt.show()
```

```python
#### draw two countries ####
# 先画China人口的折线图
plt.plot(china_pop_info['Year'], china_pop_info['Population']) 
# 再画US人口的折线图
plt.plot(us_pop_info['Year'], us_pop_info['Population']) 
# 显示画图结果
plt.show() 
```

```python
#### draw two countries(添加label参数，以示区别) ####
# 在画图之前，设置画布大小
plt.figure(figsize=(15,8)) 
plt.plot(china_pop_info['Year'], china_pop_info['Population'], label="China") 
plt.plot(us_pop_info['Year'], us_pop_info['Population'], label="United States")
# 显示图例(调整图例字体大小)
plt.legend(fontsize=16)
plt.xlabel('Year', fontsize=30)
plt.ylabel('Population Number / Billion', fontsize=20)
plt.title('China VS United States', fontsize=40)
 # 调整x轴和y轴刻度字体大小
plt.tick_params(labelsize=15)
plt.show()
```
