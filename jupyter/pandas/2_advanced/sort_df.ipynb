```python
import pandas as pd
```

```python
movie = pd.read_csv('../data/movie.csv')
(movie
   [['movie_title', 'imdb_score', 'budget']]
   .nlargest(100, 'imdb_score') 
   .nsmallest(5, 'budget')
)
```

```python
# use .sort_values to replicate the first part of the expression 
# and grab the first 100 rows with the .head method
(movie
   [['movie_title', 'imdb_score', 'budget']]
   .sort_values('imdb_score', ascending=False)
   .head(100)
)
```

```python
# now that we have the top 100 scoring movies, we can use .sort_values 
# with .head again to grab the lowest five by budget
(movie
   [['movie_title', 'imdb_score', 'budget']]
   .sort_values('imdb_score', ascending=False)
   .head(100) 
   .sort_values('budget')
   .head(5)
)
```
