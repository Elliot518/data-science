```python
import pandas as pd
```

```python
data = [["peter", "FileCode", "User", "read"],
              ["peter", "DocCode", "User", "read"],
              ["milan", "FileCode", "Manager", "read"],
              ["milan", "FileCode", "Manager", "add"],
              ["milan", "FileCode", "Manager", "edit"],
              ["milan", "FileCode", "Manager", "status"],
              ["milan", "DocCode", "Manager", "read"],
              ["milan", "DocCode", "Manager", "add"],
              ["milan", "DocCode", "Manager", "edit"],
              ["milan", "DocCode", "Manager", "status"],
              ["awink", "FileCode", "Manager", "read"],
              ["awink", "FileCode", "Manager", "add"],
              ["awink", "FileCode", "Manager", "edit"],
              ["awink", "FileCode", "Manager", "delete"],
              ["awink", "FileCode", "Manager", "status"],
              ["awink", "DocCode", "Manager", "read"],
              ["awink", "DocCode", "Manager", "add"],
              ["awink", "DocCode", "Manager", "edit"],
              ["awink", "DocCode", "Manager", "delete"],
              ["awink", "DocCode", "Manager", "status"]]
permissions = pd.DataFrame(data=data, columns=["UserName", "Project", "Role", "Permission"])
permissions
```

```python
# Custom aggregation function
def concat_permissions(permissions):
    return "/".join(sorted(permissions))

# Create pivot table with custom aggregation function
pivot = pd.pivot_table(permissions,
                       index="UserName", columns="Project",
                       values="Permission", aggfunc=concat_permissions, fill_value='')
pivot
```
